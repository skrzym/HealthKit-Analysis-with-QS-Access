```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pprint import pprint
from datetime import datetime
```

```python
# Function built to 'apply' to all date rows in the 'Start' and 'Finish' columns
def str2dt(string):
    return datetime.strptime(string,'%d-%b-%Y %H:%M')
```

```python
def int2dow(integer):
    dow_dictionary = {
    0:'Mon',
    1:'Tue',
    2:'Wed',
    3:'Thu',
    4:'Fri',
    5:'Sat',
    6:'Sun'
    }
    return dow_dictionary[integer]
```

```python
#Read the data into a pandas data frame
hd = pd.read_csv('HealthData.csv')
```

```python
#Rename the columns to simpler names for ... well simplicity!
columns = ['Start', 'Finish', 'Active_Kcal', 'Distance', 'Heart_Rate', 'Steps']
hd.columns = columns
```

```python
#Convert the 'Start' and 'Finish' columns to datetime types for easy analysis
for index in ['Start', 'Finish']:
    hd[index] = hd[index].apply(str2dt)
```

```python
#Create 6 new columns for the 'year', 'month', 'day', hour', (general) 'date', and (day of week) 'dow' values of a data point
hd['year'] = hd['Start'].apply(lambda x: x.year)
hd['month'] = hd['Start'].apply(lambda x: x.month)
hd['day'] = hd['Start'].apply(lambda x: x.day)
hd['hour'] = hd['Start'].apply(lambda x: x.hour)
hd['date'] = hd['Start'].apply(lambda x: x.date())
hd['dow'] = hd['Start'].apply(lambda x: x.weekday())
```

### 2) Visualizing Your Steps!

```python
def get_monthly_steps_per_factor(df, factor, aggfunc, year=2015, month=None):
    #First we check for a specific edge case and handle it accordingly
    if month == None:
        query = df.query('year=='+str(year))
    else:
        query = df.query('year=='+str(year)+' & month=='+str(month))
    if factor == 'dow' and aggfunc == np.mean:
        pivot = pd.pivot_table(query, index=['month','date'], values=['Steps'], aggfunc=np.sum)
        data = pivot.reset_index()
        data['dow'] = data['date'].apply(lambda x: x.weekday())
        pivot = pd.pivot_table(data, index=['month','dow'], values=['Steps'], aggfunc=np.mean)
        pivot = pivot.reset_index()
    else:
        pivot = pd.pivot_table(query, index=['month',factor], values=['Steps'], aggfunc=aggfunc)
        pivot = pivot.reset_index()
    if factor is 'dow':
        pivot['dow'] = pivot['dow'].apply(int2dow)
    return pivot
```

```python
def plot_monthly_steps_per_factor(df,factor='day',aggfunc=np.mean,year=2016,month=None,font_scale=2,col_wrap=4,size=6,aspect=1):
    mdata = get_monthly_steps_per_factor(df, factor, aggfunc, year, month)
    pprint(mdata)
    sns.set(font_scale=font_scale)
    fg = sns.FacetGrid(mdata, col='month', col_wrap=col_wrap, size=size, aspect=aspect)
    fg.map(sns.barplot,factor,'Steps')
    #Get Title Working
    #plt.gcf().suptitle('Title')
    plt.show()
```

```python
plot_monthly_steps_per_factor(df=hd, factor='hour', aggfunc=np.sum, year=2015, month=None, aspect=1.5)
```

```python
plot_monthly_steps_per_factor(df=hd, factor='dow', aggfunc=np.mean, year=2015, aspect=1)
```

```python
def get_steps_per_day(df, aggfunc):
    for month in range(1,13):
        query = df
        pivot = pd.pivot_table(query, index=['date'], values=['Steps'], aggfunc=aggfunc)
        pivot = pivot.reset_index()
    return pivot
```

```python
#Change Matplotlib's settings back to default so seaborn's formatting does not mess up the heatmap calendar
sns.reset_orig()
```

```python
def plot_heatmap_calendar(df, year, month, aggfunc, cmap='coolwarm'):
    dates, data = generate_data(df, year, month, aggfunc)
    if type(month) == int:
        fig, ax = plt.subplots()
    else:
        fig, ax = plt.subplots(figsize=(5,len(month)*4))
    calendar_heatmap(ax, dates, data, cmap)
    #Get title working
    #plt.title('Title')
    plt.show()
```

```python
def generate_data(df, year, month, aggfunc):
    query = get_steps_per_day(df.query('year=='+str(year)+' & month=='+str(month)),aggfunc)
    data = query.Steps.tolist()
    dates = query.date.tolist()
    return dates, data
```

```python
def calendar_array(dates, data):
    i, j = zip(*[d.isocalendar()[1:] for d in dates])
    i = np.array(i) - min(i)
    ####
    #Edge Case - First few days to be plotted are considered to be from the previous or following week
    if i[0] != min(i):
        if i[-1]!= min(i):
            i += 1
            max_val = max(i)
            for index in range(len(i)):
                if i[index] == max_val:
                    i[index] = 0
        else:
            print('HERE')
            min_val = min(i)
            max_val = max(i)
            for index in range(len(i)):
                if i[index] == min_val:
                    i[index] = max_val + 1
            pprint(min_val)
            pprint(max_val)
            pprint(i)
            min_val = min(i)
            i = i-min_val
    ####
    j = np.array(j) - 1
    ni = max(i) + 1

    calendar = np.nan * np.zeros((ni, 7))
    calendar[i, j] = data
    return i, j, calendar
```

```python
def calendar_heatmap(ax, dates, data, cmap):
    i, j, calendar = calendar_array(dates, data)
    im = ax.imshow(calendar, interpolation='none', cmap=cmap)
    label_days(ax, dates, i, j, calendar)
    label_months(ax, dates, i, j, calendar)
    ax.figure.colorbar(im)
```

```python
def label_days(ax, dates, i, j, calendar):
    ni, nj = calendar.shape
    day_of_month = np.nan * np.zeros((ni, 7))
    day_of_month[i, j] = [d.day for d in dates]

    for (i, j), day in np.ndenumerate(day_of_month):
        if np.isfinite(day):
            ax.text(j, i, int(day), ha='center', va='center')

    ax.set(xticks=np.arange(7), 
           xticklabels=['M', 'T', 'W', 'R', 'F', 'S', 'S'])
    ax.xaxis.tick_top()
```

```python
def label_months(ax, dates, i, j, calendar):
    month_labels = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul',
                             'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    months = np.array([d.month for d in dates])
    uniq_months = sorted(set(months))
    yticks = [i[months == m].mean() for m in uniq_months]
    labels = [month_labels[m - 1] for m in uniq_months]
    ax.set(yticks=yticks)
    ax.set_yticklabels(labels, rotation=90)
```
